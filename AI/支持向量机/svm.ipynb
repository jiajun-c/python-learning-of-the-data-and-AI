```python
from sklearn.svm import LinearSVC
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
#data = pd.DataFrame(iris)
iris.target
```

```python
feature = iris.data[:100,:2]
target = iris.target[:100]
# 只使用了两个特征
```

```python
scaler = StandardScaler()
```

```python
feature_standardized = scaler.fit_transform(feature)
```

```python
scv = LinearSVC(C=1.0)
```

```python
model = scv.fit(feature_standardized, target)
```

```python
color = ["black" if c == 0 else "lightgrey" for c in target]
```

```python
plt.scatter(feature_standardized[:,0], feature_standardized[:,1],c=color)
```

```python
w = scv.coef_[0]
a = -w[0]/w[1]
x = np.linspace(-2.5,2.5)
y = a*x - (scv.intercept_[0])/w[1]
plt.plot(x,y)
plt.axis("off")
plt.scatter(feature_standardized[:,0], feature_standardized[:,1],c=color)
plt.show()
```
